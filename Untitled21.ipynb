```python
import pandas as pd
url = "https://www.espncricinfo.com/records/highest-career-batting-average-282910"
```

```python
all_tables = pd.read_html(url)
```

```python
df = all_tables[0]
```

```python
df.head(5)
```

```python
 df =  df.rename(columns = {'NO': 'Not_out','HS': 'Highest_score','SR': 'Strike_rate','Mat':'Match','BF':'Ball_faces','Ave': 'Average'})
```

```python
df['rookie_year'] = df['Span'].str.split(pat='-').str[0]
df
```

```python
df['Final_year'] = df['Span'].str.split(pat='-').str[1]
```

```python
df.drop(['Span'], axis=1, inplace=True)
```

```python
df['Player_name'] = df['Player'].str.split(pat='(').str[0]
df['Country_name'] = df['Player'].str.split(pat='(').str[1]
df['Country_name'] = df['Country_name'].str.split(pat=')').str[0]
df.drop(['Player'], axis=1, inplace=True)
```
